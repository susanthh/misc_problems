## Define a function vowel_count() that counts and displays the number of vowels in the textfile story.txt

```python
def vowel_count(word):
    s = 0
    for i in word:
        if i in "AEIOUaeiou":
            s += 1
    return s

with open("story.txt", "r") as files:
    count = vowel_count(files.read())
    print("No. of vowels: ", count)
    
```

## Define a function countUpperLower which counts uppercase characters, lowercase characters, digits, and special characters

```python
def countUpperLower(word):
    u,l,d,s = 0,0,0,0
    for i in word:
        if i.isalpha():
            if i.isupper():
                u+=1
            elif i.islower():
                l+=1
        elif i.isdigit():
            d+=1
        else:
            s+=1
    return u,l,d,s

with open("story.txt", "r") as files:
    counts = countUpperLower(files.read())
    print("Uppercase characters: ", counts[0])
    print("Lowercase characters: ", counts[1])
    print("Digits: ", counts[2])
    print("Special characters: ", counts[3])
```

## Display all the digits in a file

```python
def displayDigits(word):
    for i in word:
        try:
            print(int(i), end = '')
        except:
            pass

with open("story.txt", "r") as files:
    displayDigits(files.read())
```

## Define a function to count occurences of 'a'

```python
def counta():
    with open("story.txt", "r") as files:
        x = files.read()
        print("Count of 'a': ", x.count("a"))
counta()

#If you want it to be case insensitive, you can do this
def counta2():
    with open("story.txt", "r") as files:
        count = 0
        x = files.read()
        for i in x:
            if i in "Aa":
                count += 1
        print("Count of 'a'(both upper and lowercase): ", count)
counta2()
```
